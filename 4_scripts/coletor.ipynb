```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# Inicializa o navegador
navegador = webdriver.Chrome()
```

```python
#Aumenta o tamanho do navegador
navegador.maximize_window()
```

```python
# Loop para capturar descrição dos produtos
Lista_produtos = []
valor_parcelado = []
valor = []
for i in range (1,8):
    navegador.get(f"https://www.kabum.com.br/hardware/placa-de-video-vga?page_number={i}&page_size=100&facet_filters=&sort=most_searched")
    for n in range(1, 101):
        try:
            Lista_produtos.append(navegador.find_element(By.XPATH, f'//*[@id="listing"]/div[3]/div/div/div[2]/div/main/div[{n}]/article/a/div/button/div/h3/span').text)
        except:
            pass
        try:
            valor.append(navegador.find_element(By.XPATH, f'/html/body/div[1]/div/div[2]/div[1]/div[3]/div/div/div[2]/div/main/div[{n}]/article/a/div/div[2]/div[2]/span').text)
        except:
            pass
        try:
            valor_parcelado.append(navegador.find_element(By.XPATH, f'/html/body/div[1]/div/div[2]/div[1]/div[3]/div/div/div[2]/div/main/div[{n}]/article/a/div/div[2]/div[3]/span/b').text)
        except:
            pass
```

```python
valor_parcelado
```

```python
#Importação para o pandas
import pandas as pd
from unidecode import unidecode
tabela1 = pd.DataFrame(Lista_produtos, columns=['Produtos'])
tabela2 = pd.DataFrame(valor, columns=['Valor'])
tabela3 = pd.DataFrame(valor_parcelado, columns=['valor parcelado'])
df = pd.concat([tabela1, tabela2, tabela3], axis=1)
```
